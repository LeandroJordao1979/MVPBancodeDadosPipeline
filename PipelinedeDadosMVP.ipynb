```python

# 1. Criar/conectar ao banco
conn = sqlite3.connect('analisecartaodecredito.db')
cursor = conn.cursor()

# 2. Criar tabelas (cole aqui os comandos CREATE TABLE)
cursor.execute("""DROP TABLE IF EXISTS DIM_CRITERIO_CLASSIFICACAO;""")
cursor.execute("""DROP TABLE IF EXISTS FATO_CREDITO;""")
cursor.execute("""DROP TABLE IF EXISTS DIM_CLIENTE;""")

cursor.execute("""
CREATE TABLE DIM_CLIENTE (
    Ind_ID INT PRIMARY KEY,
    GENDER VARCHAR(10),
    Car_Owner VARCHAR(5),
    Propert_Owner VARCHAR(5),
    Type_Income VARCHAR(50),
    EDUCATION VARCHAR(50),
    Marital_status VARCHAR(50),
    Housing_type VARCHAR(50),
    Type_Occupation VARCHAR(50)
);
""")

cursor.execute("""
CREATE TABLE FATO_CREDITO (
    Ind_ID INT,
    Annual_income FLOAT,
    Birthday_count INT,
    Employed_days INT,
    CHILDREN INT,
    Family_Members INT,
    Mobile_phone INT,
    Work_Phone INT,
    Phone INT,
    EMAIL_ID INT,
    label INT,
    FOREIGN KEY (Ind_ID) REFERENCES DIM_CLIENTE(Ind_ID)
);
""")

cursor.execute("""
CREATE TABLE DIM_CRITERIO_CLASSIFICACAO (
    ID_CRITERIO INT PRIMARY KEY,
    Descricao VARCHAR(100),
    Condicao_SQL TEXT,
    Resultado VARCHAR(20)
);
""")

conn.commit()
```
